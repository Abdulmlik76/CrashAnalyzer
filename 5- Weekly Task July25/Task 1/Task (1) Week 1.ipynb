##1. Data Cleaning

### Read the dataset

```python
import pandas as pd
df=pd.read_csv("Task (1) Dataset.csv")
df
```

### Handle Missing Values

```python
#to make na to NAN
# then fill them with mode of the column
# removed the columns with high row nulls
import numpy as np
tona=[]
df=df.drop(columns=["Defect_of_vehicle","Service_year_of_vehicle","Casualty_class","Sex_of_casualty", "Age_band_of_casualty","Casualty_severity","Fitness_of_casuality","Work_of_casuality"])
for col in df.columns:
    for x in df[col]:
        if((x=="na")|(x=="Unknown")|(x==np.nan)):
            x=np.nan
            x=df[col].fillna(df[col].dropna().mode())
            tona.append(x[0])
        else:
            tona.append(x)
    df[col]=tona
    tona.clear()
df.isna().sum()
```

```python
#dropped the minimal missing data that are relativly low to the dataframe
df=df.dropna()
# to show the resault after 
df
```

```python
#to show that there is 0 missing data
df.isna().sum()
```

### Correct any inconsistent data entries.

```python
#ALL NAN ,na and Uknown data has been fixed from above as seen here 
for col in df.columns:
    print(df[col].unique())
```

### Ensure data types are appropriate for each column.

---



```python
#the only changeable datas are these
df["Number_of_casualties"]=pd.to_numeric(df["Number_of_casualties"])

df["Number_of_vehicles_involved"]=pd.to_numeric(df["Number_of_vehicles_involved"])
```
